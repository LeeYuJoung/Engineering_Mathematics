#### **02.02.01 벡터/행렬의 덧셈과 뺄셈 (86p)**

* **요소별(Elenent-Wise) 연산**
  * 같은 크기를 가진 두 개의 벡터나 행렬에서 같은 위치에 있는 원소끼리 덧셈과 뺄셈을 하는 것

* 벡터 $x$ 또는 행렬 $A$에 스칼라값 $c$를 곱하는 것은 벡터 $x$ 또는 행렬 $A$의 모든 원소에 스칼라값 $c$를 곱하는 것과 같음

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
### 벡터의 덧셈과 뺄셈
x = np.array([10, 11, 12, 13, 14])
y = np.array([0, 1, 2, 3, 4])
```

```python
x + y
```

```python
x - y
```

```python
### 행렬의 덧셈과 뺄셈
np.array([[5, 6], [7, 8]]) + np.array([[10, 20], [30, 40]]) - np.array([[1, 2], [3, 4]])
```

* **브로드캐스팅 (Broadcasting)**
  * 원래 덧셈과 뺄셈은 같은 크기의 두 벡터에 대해서만 가능
  * 단, 벡터와 스칼라의 경우 1 -벡터를 사용하여 스칼라를 벡터로 변환하여 연산을 허용

  * $\begin{pmatrix}10\\11\\12\end{pmatrix} - 10 = \begin{pmatrix}10\\11\\12\end{pmatrix} - 10·1 = \begin{pmatrix}10\\11\\12\end{pmatrix} - \begin{pmatrix}10\\10\\10\end{pmatrix}$

* **선형조합 (Linear Combination)**
  * 벡터/행렬에 스칼라값을 곱한 후 더하거나 뺀 것을 벡터/행렬의 선형조합이라 함
  * 벡터나 행렬은 선형조합해도 ***크기는 변하지 않음***
  * $c_1x_1+c_2x_2+c_3x_3+\dots+c_Lx_L = x$

#### **02.02.02 벡터/행렬의 곱셈 (89p)**

* **내적 (Inner Product)**
  * $x⋅y = < x, y > = x^Ty$
  * 두 벡터를 내적하려면 두 가지 조건 만족해야 함
    1. 두 벡터의 차원(길이)가 같아야 함
    2. 앞의 벡터가 ***행벡터***이고, 뒤의 벡터가 ***열벡터***여야 함
  * ***내적의 결과는 스칼라값***
    * 같은 위치에 있는 원소들을 요소별로 곱한다음, 그 값들을 모두 더해 하나의 스칼라값으로 만듬

```python
### 두 2차원 벡터의 내적 계산
x = np.array([[1], [2], [3]])
y = np.array([[4], [5], [6]])

np.dot(x.T, y)
```

```python
### 두 1차원 벡터의 내적 계산
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

np.dot(x, y)
```

* **가중합 (Weighted Sum)**
  * ***벡터의 내적***은 가중합을 계산할 때 쓰임
  * 가중합이란 복수의 데이터를 단순히 합하는 것이 아니라 각각의 수에 ***어떤 가중치를 곱한 후 이 곱셈 결과들을 다시 합***한 것을 뜻함
  * 예) 데이터 벡터가 $x = [x_1, \dots x_N]^T$이고, 가중치 벡터가 $w = [w_1, \dots w_N]^T$일 때 데이터의 가중합

    ➜  $w_1x_1 + \dots + w_Nx_N = \displaystyle\sum_{i=1}^{N} w_ix_i$

    ➜ $w^Tx \quad or \quad x^Tw$

```python
### 연습문제) 02.02.01
# 데이터 벡터
p = np.array([100, 80, 50])
# 가중치 벡터
n = np.array([3, 4, 5])

np.dot(p, n)
```

* **가중평균 (Weighted Average)**
  * 가중합의 ***가중치값을 전체 가중치값의 합으로 나누면*** 가중평균이 된다.


* 벡터로 표현된 $n$개의 단순 평균
  * $x$ 데이터의 평균은 보통 $\overline{x}$ 라는 기호로 표기
  
    $\overline{x}  = \frac{1}{N}\displaystyle\sum_{i=1}^{N} x_i = \frac{1}{N}1^T_Nx$

```python
### 평균 계산
x = np.arange(10)
N = len(x)

(np.ones(N) @ x) / N
# np.dot(np.ones(N), x) / N
```

```python
x.mean()
```

* **유사도 (Similarity)**
  * 벡터의 내적은 두 벡터 간의 유사도를 계산하는데 이용할 수도 있음
  * 유사도는 ***두 벡터가 닮은 정도***를 정량적으로 나타낸 값
  * 두 벡터가 비슷한 경우 유사도 값이 커지고, 비슷하지 않은 경우 유사도 값이 작아짐
  * 내적을 이용하면 ***코사인 유사도***라는 유사도를 계산 가능

```python
### 0과 1을 나타내는 MNIST 이미지에 대한 내적 계산
from sklearn.datasets import load_digits
import matplotlib.gridspec as gridspec
import matplotlib.pylab as plt
import numpy as np

digits = load_digits()
d1 = digits.images[0]
d2 = digits.images[10]
d3 = digits.images[1]
d4 = digits.images[11]
v1 = d1.reshape(64, 1)
v2 = d2.reshape(64, 1)
v3 = d3.reshape(64, 1)
v4 = d4.reshape(64, 1)

plt.figure(figsize=(9, 9))
gs = gridspec.GridSpec(1, 8, height_ratios=[1], width_ratios=[9, 1, 9, 1, 9, 1, 9, 1])

for i in range(4):
  plt.subplot(gs[2*i])
  plt.imshow(eval('d' + str(i + 1)), aspect = 1, interpolation = 'nearest', cmap = plt.cm.bone_r)
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  plt.title('Image {}'.format(i + 1))
  plt.subplot(gs[2 * i + 1])
  plt.imshow(eval('v' + str(i + 1)), aspect = 0.25, interpolation = 'nearest', cmap = plt.cm.bone_r)
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  plt.title('Vector {}'.format(i + 1))

plt.tight_layout()
plt.show()
```

```python
np.dot(v1.T, v2), np.dot(v3.T, v4)
```

```python
np.dot(v1.T, v3), np.dot(v1.T, v4), np.dot(v2.T, v3), np.dot(v2.T, v4)
```

#### **02.02.03 선형회귀 모형 (95p)**

* **선형회귀 모형 (Linear Regression Model)**
  * ***독립변수 $x$에서 종속변수 $y$를 예측하는 방법***의 하나로 독립변수 벡터 $x$와 가중치 벡터 $w$와의 가중합으로 $y$에 대한 ***예측값 $\hat{y}$ (와이 햇(hat))를 계산하는 수식***
  * $\hat{y} = w_1x_1 + \dots + w_Nx_N = w^Tx$
  * 선형회귀 모형은 비선형적인 현실 세계의 데이터를 잘 예측하지 못할 수 있다는 단점 존재

* **제곱합 (Sum of Squares)**
  * 데이터의 분산이나 표준편차등을 구하는 경우에는 각각의 데이터를 제곱한 뒤 이 값을 모두 더한 제곱합을 계산해야 함
  * 이 경우에도 벡터의 내적을 사용 가능 ➜ $x^Tx$

* 행렬의 곱셈은 곱하는 행렬의 순서를 바꾸는 ***교환 법칙 성립하지 않음***
* 덧셈에 대한 ***분배 법칙은 성립***

  $AB \neq BA$

  $A(B+C) = AB + AC$

  $(A+B)C = AC + BC$

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.array([[9, 8], [7, 6]])
```

```python
## 교환 법칙 성립 불가
np.dot(A, B)
```

```python
np.dot(B, A)
```

```python
## 분배 법칙 성립
np.dot(A, (B+C))
```

```python
np.dot(A, B) + np.dot(A, C)
```

#### **02.02.06 잔차 (108p)**

* 선형회귀 분석한 결과는 가중치 벡터 $w$라는 형태로 나타나고, 예측치는 이 가중치 벡터를 사용한 독립변수 데이터 레코드 즉, 벡터 $x_i$의 가중합 $w^Tx_i$가 됨
* 예측치와 실젯값 $y_i$의 차이를 ***오차 혹은 잔차 $e_i$***라고 함
* 이러한 잔찻값을 모든 독립변수 벡터에 대해 구하면 잔차벡터 $e$가 된다

  $e_i = y_i - \hat{y}_i = y_i - w^Tx_i$

* **이차형식**
  * 어떤 벡터와 정방행렬이 ***'행벡터 X 정방행렬 X 열벡터'*** 형식으로 되어 있는 것

  * $x^TAx = \displaystyle\sum_{i=1}^{N}\displaystyle\sum_{j=1}^{N} a_{i,j}x_ix_j$

```python
### 연습문제) 02.02.11
# 이차형식 계산
x = np.array([1, 2, 3])
A = np.arange(1, 10).reshape(3, 3)
x.T @ A @ x
```
